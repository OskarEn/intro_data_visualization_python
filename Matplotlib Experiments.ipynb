```python
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3], [1,4,9])
plt.title("Title for my simple graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show
```

```python
plt.plot([1,2,3], [1,4,9])
plt.plot([1.5,3,4], [2,6,10])
plt.legend(['set 1', 'set 2'])
plt.title('Title for my simple graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show
plt.savefig('saved-image')
```
