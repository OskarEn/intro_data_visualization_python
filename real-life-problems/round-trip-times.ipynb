```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
distance = [9395, 5204, 2851]
avg_time = [239.795, 201.371, 106.924]
plt.scatter(avg_time, distance)
plt.xlabel("time (ms)")
plt.ylabel("distance (km)")
plt
plt.show()
```

So we have the points. There are loads of ways to do the line; three that I found:

    - https://stackoverflow.com/a/37294651
- scikits.statsmodels
- seaborn.regplot 

```python
import seaborn as sb
```

```python
plt.xlim(0, 250)
plt.ylim(0, 10000)
plt.title("RTT with three points of data")
plt.xlabel("average time (ms)")
plt.ylabel("distance (km)")
sb.regplot(avg_time, distance, scatter=True, fit_reg=True, ci=95)
# plt.show()
plt.savefig("rtt-plot-95-accuracy")
```

```python
## It would be good to plot the standard deviation values for each measurement. Umtil then, just having the line drawn is cleaner
plt.xlim(0, 250)
plt.ylim(0, 10000)
plt.title("RTT by Pinging Universities in 3 Different Countries")
plt.xlabel("average time (ms)")
plt.ylabel("distance (km)")
sb.regplot(avg_time, distance, scatter=True, fit_reg=True, ci=False)
# plt.show()
plt.savefig("rtt-plot")
```
