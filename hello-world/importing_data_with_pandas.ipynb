```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Importing data
data = pd.read_csv('countries.csv')
data.head #success!
```

```python
#Now to gain insights into, say, the GDP development of Afghanistan...
afghanistan = data[data.country == 'Afghanistan']
plt.plot(afghanistan.year, afghanistan.gdpPerCapita)
plt.title('Afghanistan GDP')
plt.xlabel('Year')
plt.ylabel('GDP per Capita USD')
plt.show
plt.savefig('afghanistan-economic-data')
```
