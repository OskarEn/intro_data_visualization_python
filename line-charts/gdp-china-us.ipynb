```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("countries.csv")
data.head
```

```python
# How has the GDP in the US changed
us_gdp = data[data.country=='United States']
#us_gdp.head
plt.plot(us_gdp.year, us_gdp.gdpPerCapita)
plt.xlabel("Year")
plt.ylabel("USD")
plt.title("US GDP per Capita")
```

```python
# China
china_gdp = data[data.country=="China"]
china_gdp.head
plt.plot(china_gdp.year, china_gdp.gdpPerCapita)
plt.xlabel("Year")
plt.ylabel("USD")
plt.title("China GDP per Capita")
```

```python
# We can plot both by calling plt.plot(..) multiple times
plt.plot(us_gdp.year, us_gdp.gdpPerCapita)
plt.plot(china_gdp.year, china_gdp.gdpPerCapita)
plt.xlabel("Year")
plt.ylabel("USD")
plt.legend(["US", "China"])
plt.show()
```

```python
# It is however more helpful to look at the relative growth. We will use the first year collected as our index
# iloc = integer location
us_index = us_gdp.gdpPerCapita.iloc[0]
us_growth = us_gdp.gdpPerCapita / us_index * 100

china_index = china_gdp.gdpPerCapita.iloc[0]
china_growth = china_gdp.gdpPerCapita / china_index * 100

plt.plot(us_gdp.year, us_growth)
plt.plot(china_gdp.year, china_growth)
plt.title("")
plt.xlabel("Year")
plt.ylabel("USD")
plt.legend(["US", "China"])
plt.savefig("relative-gdp-growth-china-us")
```
