```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("countries.csv")
```

```python
us = data[data.country == "United States"]
china = data[data.country == "China"]
# us.head
# china.head
```

```python
# Absolute
plt.plot(us.year, us.population)
plt.plot(china.year, china.population)
plt.legend(["US", "China"])
plt.title("Population")
```

```python
# Relative
us_relative = us.population / us.population.iloc[0] * 100
china_relative = china.population / china.population.iloc[0] * 100

plt.plot(us.year, us_relative)
plt.plot(china.year, china_relative)
plt.legend(["US", "China"])
plt.title("Population Growth since start")
```
