```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data_big = pd.read_csv("obama_too_big.csv", parse_dates=["year_month"])
data = pd.read_csv("obama.csv", parse_dates=["year_month"])
```

```python
# Size of the data
print("Big data " + str(data_big.shape))
print("Normal data " + str(data.shape))
```

```python
# plotting data_big 

plt.plot(data_big.year_month, data_big.approve_percent, "o", markersize=1, alpha=0.2)
plt.show()
```

```python
# to take a sample of this data and get a clearer view
# .sample(n=xxx)
# .sample(frac=0.x)
tenth_sample = data_big.sample(frac=0.1)
plt.plot(tenth_sample.year_month, tenth_sample.approve_percent, "o", markersize=1, alpha=0.2)
plt.show()
```
